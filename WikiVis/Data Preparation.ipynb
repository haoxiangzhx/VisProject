```python
# select possible qualified visualization questions
def filter_questions(part_file_name):
    with open('../WikiSQL/data/'+part_file_name+'.jsonl') as f:
        questions_list = f.readlines()

    filtered_questions_list = []
    for question in questions_list:
        question_json = json.loads(question)
        conds = question_json['sql']['conds']
        if len(conds) == 1 and conds[0][1] != 0:
            if question_json['sql']['agg'] >= 3:
                filtered_questions_list.append(question_json)
            
    return filtered_questions_list
```

```python
def get_tables(part_file_name):
    with open('../WikiSQL/data/'+part_file_name+'.tables.jsonl') as f:
        tables_list = f.readlines()
        
    tables = {}
    for table in tables_list:
        table_json = json.loads(table)
        tables[table_json['id']] = table_json
    return tables
```

```python
filtered_train_questions = filter_questions('train')
filtered_dev_questions = filter_questions('dev')
filtered_test_questions = filter_questions('test')
```

```python
train_tables_dict = get_tables('train')
dev_tables_dict = get_tables('dev')
test_tables_dict = get_tables('test')
```

```python
# show, what, how
def modify_questions(questions, tables_dict):
    new_questions = []
    for question in questions:
        table_id = question['table_id']
        table = tables_dict[table_id]
        header = table['header']
        sql = question['sql']
        sel = sql['sel']
        sel_col = header[sel]
        conds = sql['conds'][0][0]
        conds_col = header[conds]
        agg = sql['agg']
        if agg == 3:
            question['question1'] = 'What is the '+sel_col+' over different '+conds_col
            question['question2'] = 'How many '+sel_col+' over different '+conds_col
        elif agg == 4:
            question['question1'] = 'What is the sum of '+sel_col+' over different '+conds_col
            question['question2'] = 'How many '+sel_col+' over different '+conds_col
        else:
            question['question1'] = 'What is the average value of '+sel_col+' over different '+conds_col
            question['question2'] = 'How does the average '+sel_col+' distribute over different '+conds_col
        new_questions.append(question)
    return new_questions
```

```python
new_train_questions = modify_questions(filtered_train_questions, train_tables_dict)
new_dev_questions = modify_questions(filtered_dev_questions, dev_tables_dict)
new_test_questions = modify_questions(filtered_test_questions, test_tables_dict)

file_names = ['new_train_questions', 'new_dev_questions', 'new_test_questions']
new_questions = [new_train_questions, new_dev_questions, new_test_questions]
for file_name, new_question in zip(file_names, new_questions):
    with open(file_name+'.txt', 'w') as outfile:
        for question in new_question:
            json.dump(question, outfile)
            outfile.write('\n')
            
# Then manually check, finish 1st 50 Saturday night
# Tell me the average Laps for different grid
# show, what, how
```
